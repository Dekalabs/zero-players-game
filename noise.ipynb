```python
import numpy as np

WIDTH = 1024
HEIGHT = 600
BLOCK = 10


tile = np.empty((WIDTH // BLOCK, HEIGHT // BLOCK))
```

```python
import opensimplex


opensimplex.seed(1234)
```

```python
for x in range(WIDTH // BLOCK):
    for y in range(HEIGHT // BLOCK):
        tile[x, y] = opensimplex.noise2(x, y)
```

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
im = ax.imshow(tile.T)
plt.show()
```
